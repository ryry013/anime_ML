```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

anime = pd.read_csv('../data/raw/svano_db/anime.csv', sep='\t', keep_default_na=False)

# Convert these columns from str to numeric
anime.score = pd.to_numeric(anime.score, downcast='float')
anime.members_count = pd.to_numeric(anime.members_count, downcast='integer')

print(len(anime))
```

```python
x = []
y = []

for i, s in enumerate(anime.score):
    if not np.isnan(s):
        if anime.status.iloc[i] == 'Finished Airing':
            x.append(s)
            y.append(anime.members_count.iloc[i])

print(len(x), len(y))
```

```python
plt.scatter(x, y)
plt.xlabel("Show rating")
plt.ylabel("Show popularity")
plt.title("MyAnimeList Popularity vs Rating")

plt.show()
```
